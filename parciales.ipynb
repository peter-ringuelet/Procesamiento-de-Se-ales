```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Definimos la función triangular (tri(f/2))
def triangular(f):
    return np.maximum(1 - np.abs(f), 0)

# Definimos la función rectangular (rect(f/3))
def rectangular(f):
    return np.where(np.abs(f) <= 1.5, 1/3, 0)

# Definimos el dominio de la frecuencia
f = np.linspace(-5, 5, 500)

# Calculamos las funciones
triangular_f2 = triangular(f / 2)
rectangular_f3 = rectangular(f / 3)

# Calculamos la convolución de las dos funciones
convolucion = convolve(triangular_f2, rectangular_f3, mode='same')

# Graficamos las funciones y su convolución
plt.figure(figsize=(10, 6))

# Función triangular
plt.subplot(3, 1, 1)
plt.plot(f, triangular_f2, label="tri(f/2)", color='blue')
plt.title("Función triangular tri(f/2)")
plt.grid(True)

# Función rectangular
plt.subplot(3, 1, 2)
plt.plot(f, rectangular_f3, label="rect(f/3)", color='green')
plt.title("Función rectangular rect(f/3)")
plt.grid(True)

# Convolución de las dos funciones
plt.subplot(3, 1, 3)
plt.plot(f, convolucion, label="Convolución", color='red')
plt.title("Convolución de tri(f/2) y rect(f/3)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
