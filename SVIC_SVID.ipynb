```python
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    """
    Genera la señal h(t) del ejercicio 2 de la práctica.
    """
    return (t + 1) * ((t >= -1) & (t < 1)) + (-t + 3) * ((t >= 1) & (t < 3))

# Definir el vector de tiempo
t = np.arange(-2, 8, 0.001)

# Evaluar la señal original h(t)
h_original = x(t)

# Evaluar la señal desplazada h(2t - 3)
h_desplazada = x(2 * t - 3)

# Graficar ambas señales
plt.plot(t, h_original, label='h(t)')
plt.plot(t, h_desplazada, label='h(2t - 3)', linestyle='--')
plt.title('Señal original h(t) y señal desplazada h(2t - 3)')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def x(n):
    return np.where((n >= 0) & (n <= 4), 1, 0)

# Definir el rango de n
n = np.arange(-5, 10)

# Graficar señal usando un stem plot
plt.stem(n, x(n), label='x[n]')


# Configuración de la gráfica
plt.title('Señal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()

```

```python
# Definir la función x(t)
def x(t):
    return -2 * np.sin(-0.2 * t + 5 * np.pi / 3)

# Definir el rango de t
t = np.linspace(0, 100, 1000)

# Evaluar la función x(t)
y = x(t)

# Graficar la señal
plt.plot(t, y, label=r'$x(t) = -2 \sin(-0.2t + \frac{5\pi}{3})$')
plt.title(r'$x(t) = -2 \sin(-0.2t + \frac{5\pi}{3})$')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.yticks(ticks=[-2,-1,0,1,2,3])

plt.legend()

# Mostrar la gráfica
plt.show()


# Definir el rango de n
n = np.linspace(0, 100, 1000)
# Graficar señal usando un stem plot
plt.stem(n, x(n), label=r'$x[n] = -2 \sin(-0.2t + \frac{5\pi}{3})$')
plt.title(r'$x[n] = -2 \sin(-0.2t + \frac{5\pi}{3})$')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.yticks(ticks=[-2,-1,0,1,2,3])

plt.legend()

# Mostrar la gráfica
plt.show()
```
