```python
from oct2py import octave
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Número de alumno
numAl = 29607

# Llama a la función senial.m desde Octave
n, x = octave.senial(numAl)

# Convertir los resultados a arrays de NumPy
n = np.array(n)
x = np.array(x)

# Gráfico de la señal discreta
plt.figure(1)
plt.stem(n, x, basefmt=" ", use_line_collection=True)
plt.title("Señal discreta (03775/3)")
plt.xlabel("n")
plt.ylabel("x[n]")

# Guardar la gráfica como imagen .png
plt.savefig("GraficoSenial.png")

# Mostrar el gráfico en pantalla
plt.show()
```
